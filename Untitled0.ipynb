```python
import pandas as pd

df = pd.read_csv("sample.csv")
text_data = df["text"]

print(text_data.head())
```

**13.Write code to convert a text string to lowercase.**

```python
df['text_lower'] = df['text'].str.lower()
df[['text', 'text_lower']].head()
```

14.Write a Python function to remove punctuation from text.

```python
import string

df['no_punctuation'] = df['text'].apply(
    lambda x: x.translate(str.maketrans('', '', string.punctuation))
)

df[['text', 'no_punctuation']].head()
```

**15.Implement stopword removal using NLTK.**

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Added to address the LookupError

stop_words = set(stopwords.words('english'))

df['no_stopwords'] = df['text'].apply(
    lambda x: " ".join([w for w in word_tokenize(x) if w.lower() not in stop_words])
)

df[['text', 'no_stopwords']].head()
```

16.Perform stemming using Porter Stemmer on a sample sentence

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

df['stemmed'] = df['text'].apply(
    lambda x: " ".join([ps.stem(w) for w in word_tokenize(x)])
)

df[['text', 'stemmed']].head()
```

17.Perform lemmatization using spaCy or WordNet

```python
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()

df['lemmatized'] = df['text'].apply(
    lambda x: " ".join([lemmatizer.lemmatize(w) for w in word_tokenize(x)])
)

df[['text', 'lemmatized']].head()

```

18.Write regex code to remove URLs from text

```python

import pandas as pd
import re


df = pd.read_csv("sample.csv")


df['no_urls'] = df['text'].apply(
    lambda x: re.sub(r'http\\S+|www\\S+', '', x)
)

df[['text', 'no_urls']].head()

```

19. Write code to remove HTML tags.

```python
import pandas as pd
import re


df = pd.read_csv("sample.csv")

df["no_html"] = df["text"].apply(
    lambda x: re.sub(r"<.*?>", "", x) if isinstance(x, str) else x
)

print(df[["text", "no_html"]].head())

```
